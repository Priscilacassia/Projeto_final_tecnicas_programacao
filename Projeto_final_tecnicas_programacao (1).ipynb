```python
#Retirando espaços e padronizando o nome das colunas.
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
#Trocando vírgula por ponto
df = df.replace(',', '.', regex=True)
```

Questões:

```python
#1.Qual é a expectativa de vida média global?
print(df.life_expectancy.mean())
```

```python
#2. Em quais países desenvolvidos vivem mais?
df.groupby('status', as_index=False)['life_expectancy'].mean()
#Países desenvolvidos vivem, em média, 12 anos a mais
#diferença muito significativa.
```

```python
#3.Faça uma comparação envolvendo mínimo, média e máximo.
df.groupby('status')['life_expectancy'].agg(['mean', 'min', 'max'])
#A análise por grupo mostrou que países desenvolvidos apresentam maior expectativa
#de vida média (79,19 anos) em comparação aos países em desenvolvimento (67,11 anos).
#Além disso, os países em desenvolvimento apresentam valores mínimos significativamente
#menores (36,3 anos), evidenciando condições mais críticas em alguns casos.
#Apesar disso, ambos os grupos podem atingir valores máximos semelhantes (89 anos),
#indicando que países em desenvolvimento também podem alcançar altos níveis de longevidade.
#Os resultados evidenciam desigualdade global na qualidade de vida e acesso à saúde.
```

```python
#4.Calcular a média da expectativa de vida no Brasil.
df[df['country'] == 'Brazil'].groupby('status')['life_expectancy'].mean()
#Ao analisar especificamente o Brasil, observa-se uma expectativa de vida média
#de 73,38 anos, valor superior à média dos países em desenvolvimento,
#indicando uma posição intermediária no cenário global.
```

```python
#5.Quais países têm maior expectativa de vida?
df.groupby('country')['life_expectancy'] \
  .mean() \
  .sort_values(ascending=False) \
  .head(10)
```
